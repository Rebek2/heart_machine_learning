```python
fig1, axes = plt.subplots(2, 3, figsize=(18, 10))
fig1.suptitle('Podstawowa analiza danych', fontsize=16)
g = sns.lineplot(x='Age',y = 'Income',data=df,hue = 'Sex', ci=None,ax=axes[0,0])
g.legend(labels=['Kobieta','Mezczyzna'])

g1=sns.lineplot(x='Age',y = 'Income',data=df,hue = 'HeartDiseaseorAttack', ci=None,ax=axes[0,1])
g1.legend(labels = ['Brak ataku','Atak serca'])

g2 = sns.lineplot(x='Age',y = 'BMI',data=df,hue = 'HeartDiseaseorAttack', ci=None,ax=axes[0,2])
g2.legend(labels = ['Brak ataku','Atak serca'])

g3 = sns.lineplot(x='Age',y = 'GenHlth',data=df,hue = 'HeartDiseaseorAttack', ci=None,ax=axes[1,0])
g3.legend(labels = ['Brak ataku','Atak serca'])

g4 = sns.lineplot(x='Age',y = 'Income',data=df,hue = 'Stroke', ci=None,ax=axes[1,1])
g4.legend(labels = ['Brak udaru','udar'])

g5 = sns.lineplot(x='Age',y = 'BMI',data=df,hue = 'Stroke', ci=None,ax=axes[1,2])
g5.legend(labels = ['Brak udaru','udar'])

plt.savefig('Plots/lineplots.png')
```

```python
sns.boxplot(x='DiffWalk',y='BMI',data=df,hue='Sex')
plt.legend(labels=['Kobieta','Mezczyzna'])
plt.xlabel('Czy problemy z chodzeniem')
```

```python
fig2, axes = plt.subplots(2, 3, figsize=(18, 10))
fig2.suptitle('Boxploty dla danych', fontsize=16)
sns.boxplot(ax=axes[0, 0], data=df, x='Sex', y='BMI')
sns.boxplot(ax=axes[0, 1], data=df, x='PhysActivity', y='BMI',hue='Sex')
sns.boxplot(ax=axes[0, 2], data=df, x='Fruits', y='BMI')
sns.boxplot(ax=axes[1, 0], data=df, x='HighChol', y='BMI')
sns.boxplot(ax=axes[1, 1], data=df, x='MentHlth', y='BMI')
sns.boxplot(ax=axes[1, 2], data=df, x='Diabetes', y='BMI')
plt.savefig('Plots/boxplots.png')
```

```python
fig3, axes = plt.subplots(2, 3, figsize=(18, 10))
fig3.suptitle('Histogramy dla danych', fontsize=16)
sns.histplot(ax=axes[0, 0],data=df,x='BMI',binwidth=3)
sns.histplot(ax=axes[0, 1],data=df,x='Income',binwidth=1)
sns.histplot(ax=axes[0, 2],data=df,x='Age',binwidth=1)
sns.histplot(ax=axes[1, 0],data=df,x='Education',binwidth=1)
sns.histplot(ax=axes[1, 1],data=df,x='GenHlth',binwidth=1)
sns.histplot(ax=axes[1, 2],data=df,x='MentHlth',binwidth=2)
#zapisac subploty (1 komenda)
#zrobic krotka prezentacje o danych na podstawie wykresow

plt.savefig('Plots/hist.png')
```

```python
fig4, axes = plt.subplots(2, 3, figsize=(18, 10))
fig4.suptitle('Podstawowa analiza danych - liczebność i relacje danych', fontsize=16)
g = sns.countplot(x="Diabetes", hue="Sex",orient ='vertical', data=df,ax=axes[0,0],palette = 'magma')
g.legend(labels=['Kobieta','Mezczyzna'])

g1 = sns.countplot(y="Age", hue="HeartDiseaseorAttack", data=df,ax=axes[0,1],palette = 'flare')
g1.legend(labels=['Brak ataku serca','Atak serca'])

g2 = sns.countplot(y="Age", hue="Smoker", data=df,ax=axes[0,2],palette = 'pastel')
g2.legend(labels=['Nie palący','Palący'])

g3 = sns.countplot(y="GenHlth", hue="Diabetes", data=df,ax=axes[1,0],palette = "light:#5A9")
g3.legend(labels=['Brak cukrzycy','Cukrzyca typu I','Cukrzyca typu II'])

g4 = sns.countplot(x="HighChol", hue="HeartDiseaseorAttack", data=df,ax=axes[1,1],palette = 'husl')
g4.legend(labels=['Brak ataku serca','Atak serca'])

g5 = sns.countplot(x="Diabetes", hue="HeartDiseaseorAttack", data=df,ax=axes[1,2],palette = 'Set2')
g5.legend(labels=['Brak ataku serca','Atak serca'])

plt.savefig('Plots/countplot.png')
```
