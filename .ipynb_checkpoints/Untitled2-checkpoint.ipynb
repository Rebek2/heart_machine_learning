```python
import csv
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015.csv")
                 
```

```python
df.describe()
```

```python

g = sns.lineplot(x='Age',y = 'Income',data=df,hue = 'Sex', ci=None)
g.legend(labels=['Kobieta','Mezczyzna'])
plt.show(g)
```

```python
sns.boxplot(x='DiffWalk',y='BMI',data=df,hue='Sex')
plt.legend(labels=['Kobieta','Mezczyzna'])
plt.xlabel('Czy problemy z chodzeniem')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Podstawowa analiza danych')
sns.boxplot(ax=axes[0, 0], data=df, x='Sex', y='BMI')
sns.boxplot(ax=axes[0, 1], data=df, x='PhysActivity', y='BMI',hue='Sex')
sns.boxplot(ax=axes[0, 2], data=df, x='Fruits', y='BMI')
sns.boxplot(ax=axes[1, 0], data=df, x='HighChol', y='BMI')
sns.boxplot(ax=axes[1, 1], data=df, x='MentHlth', y='BMI')
sns.boxplot(ax=axes[1, 2], data=df, x='Diabetes', y='BMI')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Histogramy')
sns.histplot(ax=axes[0, 0],data=df,x='BMI',binwidth=3)
sns.histplot(ax=axes[0, 1],data=df,x='Income',binwidth=1)
sns.histplot(ax=axes[0, 2],data=df,x='Age',binwidth=1)
sns.histplot(ax=axes[1, 0],data=df,x='Education',binwidth=1)
sns.histplot(ax=axes[1, 1],data=df,x='GenHlth',binwidth=1)
sns.histplot(ax=axes[1, 2],data=df,x='MentHlth',binwidth=2)
```
