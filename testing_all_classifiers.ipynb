```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import classification_report, confusion_matrix
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015.csv")
```

```python
clas = [
    LinearSVC(class_weight = 'balanced',random_state = 0),
    KNeighborsClassifier(),
    SVC(kernel ='poly' ,class_weight = 'balanced',random_state = 0),
    SVC(kernel ='linear' ,class_weight = 'balanced',random_state = 0),
    SVC(kernel ='sigmoid' ,class_weight = 'balanced',random_state = 0),
    DecisionTreeClassifier(class_weight = 'balanced',random_state = 0),
    RandomForestClassifier(class_weight = 'balanced',random_state = 0),
    AdaBoostClassifier(random_state = 0),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
    GradientBoostingClassifier(loss = 'deviance',random_state = 0),
    GradientBoostingClassifier(loss = 'exponential',random_state = 0),
]

#kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} or callable, default=’rbf’
#loss{‘log_loss’, ‘deviance’, ‘exponential’}, default=’log_loss’ GBC
```

```python
X=df.drop('HeartDiseaseorAttack',axis=1)
y = df['HeartDiseaseorAttack']
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle = True,test_size = 0.1,train_size = 0.1,
                                                     stratify=y,random_state=0)

klasyf = 0
for i in clas:
    klasyf = i
    klasyf.fit(X_train,y_train)
    y_pred = klasyf.predict(X_test)
    print('{}:'.format(i))
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))#print zrobic
    print('------------------------------------------')
    print('------------------------------------------')
```
