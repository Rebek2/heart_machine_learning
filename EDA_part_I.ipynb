```python
import csv
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
rows = []
with open("heart_disease_health_indicators_BRFSS2015.csv", 'r') as file:
    csvreader = csv.reader(file)
    headers = next(csvreader)
    for row in csvreader:
        rows.append(row)
```

```python
for i in range(len(rows)):
    for j in range(len(headers)):
        rows[i][j] = float(rows[i][j])
        
```

```python
highBPsmoke = 0

for row in rows:
    if row[1] == 1.0 and row[5] == 1.0:
        highBPsmoke = highBPsmoke + 1       
          
highBPsmoke
```

```python
highBP = 0
for row in rows:
    if row[1] == 1.0:
        highBP = highBP + 1
        
highBP
```

```python

bmi = 0
for row in rows:
    if row[1] == 1.0:
        if float(row[4]) > 25:
            bmi = bmi + 1
        

activity = 0
for row in rows:
    if row[1] == 1.0 and row[8]==1.0:

            activity = activity + 1

examined = 253680
        
```

```python
height = [examined,highBP,highBPsmoke,bmi,activity]
bars = ['Ilosc wszystkich osob','Ilosc osob z wysokim BP','Ilosc osob\n palacych z wysokim BP ','Ilosc osob z \nwysokim BP \nprzy BMI>25','Ilosc osob\n z wysokim BP\n aktywnych fizycznie']
y_pos = np.arange(len(height))
plt.figure(figsize=(15,10))
 

plt.bar(y_pos, height, color = 'black')
  

plt.xticks(y_pos, bars)
 
plt.xlabel('Typ', fontsize=25, color='#323232')
plt.ylabel('Ilosc osob', fontsize=25, color='black')
plt.title('Ilosc osob z wyosokim pulsem w zaleznosci od okolicznosci', fontsize=30, color='#323232')
 

# Show graphic
plt.show();
```
